Trường object_cols. Xếp loại theo cột

```python
object_cols = [f for f in ttgiang_read_csv.columns if ttgiang_read_csv[f].dtype =="O"]
print(object_cols)
```

```python
for col in object_cols:
  ttgiang_read_csv[col].value_counts().plot(kind='bar', figsize=(15,5))
  plt.title(col)
  plt.show()
```

Trường Numberic_cols

```python
numeric_cols = [f for f in ttgiang_read_csv.columns if ttgiang_read_csv[f].dtype !="O"]
print(numeric_cols)
```

```python
for col in numeric_cols:
  ttgiang_read_csv[col].hist()
  plt.title(col)
  plt.show()
```

```python
for col in numeric_cols:
  ttgiang_read_csv.boxplot(column=[col])
  plt.title(col)
  plt.show()
```
